```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from skimage import io, img_as_float
from skimage.filters import gaussian 
```

```python
img_gaussian_noise = img_as_float(io.imread("image.jpg", as_gray=True))
```

```python
img = img_gaussian_noise
```

```python
gaussian_kernel = np.array([[1/16, 1/8, 1/16], 
                            [1/8, 1/4, 1/8], 
                            [1/16, 1/8, 1/16]])
```

```python
conv_using_cv2 = cv2.filter2D(img, -1, gaussian_kernel, borderType=cv2.BORDER_CONSTANT)
```

```python
gaussian_using_cv2 = cv2.GaussianBlur(img, (3,3), 0, borderType=cv2.BORDER_CONSTANT)
```

```python
gaussian_using_skimage = gaussian(img, sigma=2, mode='constant', cval=0.0)
```

```python
plt.imshow(img)
plt.show()
```

```python
plt.imshow(conv_using_cv2)
plt.show()
```

```python
plt.imshow(gaussian_using_cv2)
plt.show()
```

```python
plt.imshow(gaussian_using_skimage)
plt.show()
```
